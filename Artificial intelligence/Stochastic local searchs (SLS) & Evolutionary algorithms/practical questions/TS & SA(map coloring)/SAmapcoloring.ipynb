```python
import numpy as np
import random
import math
import pandas as pd

def read_neighborhoods():
    temp= pd.read_excel('./ProvinceNeighbours.xlsx')
    np_array=temp.to_numpy()
    return np_array

def time_temprature_schedule(time,T):
    return 1/(2*time+1)*T
    #return T/math.log2(2+time)

def SA_for_map_coloring(neighborhoods,num_colors):
# SA patameters
    T=100
    initial_T=100
    coefficient=0.9999 # 0.999 is better than 0.99 and 0.9999 is bether than 0.999 and so on ... but increases the runtime
    time=0
    k=num_colors
    best_so_far_value=0
    best_so_far_solution=[]
# initial solution
    current=[]
    for i in range(31):
        current.append(random.randint(0,k-1))
    current_value=0
    for i in range(31):
        for j in range(31):
            if i>=j and neighborhoods[i][j+1]==1:
                if current[i]==current[j]:
                    current_value+=1
    best_so_far_value=current_value
    best_so_far_solution=current
#Sa
    while(1):
        #T = time_temprature_schedule(time,initial_T)
        time+=1
        T*=coefficient
        if T < 0.1 or best_so_far_value == 0:
            return best_so_far_solution,best_so_far_value
        # construct a neighbor
        next=current
        next[random.randint(0,30)]=random.randint(0,k-1)
    
        # evaluation neighbor
        next_value=0
        for g in range(31):
            for h in range(31):
                if g>=h and neighborhoods[g][h+1]==1:
                    if next[g]==next[h]:
                        next_value+=1
        #Delta E
        deltaE=next_value-current_value

        if deltaE<0:
            current=next
            current_value=next_value
            if current_value<best_so_far_value:
                best_so_far_value=current_value
                best_so_far_solution=current
        else:
            if random.random()<= math.exp(-deltaE/T):
                current=next
                current_value=next_value

neighbors=read_neighborhoods()

best_solution,minimum_conflicts=SA_for_map_coloring(neighbors,3)
print("If we use 3 colors, minimum conflicts are : ",minimum_conflicts)
print("Solution is: ")
for i in range(31):
    print(neighbors[i][0],": color ", best_solution[i])




print()

best_solution,minimum_conflicts=SA_for_map_coloring(neighbors,4)
print("If we use 4 colors, minimum conflicts are : ",minimum_conflicts)
print("Solution is: ")
for i in range(31):
    print(neighbors[i][0],": color ", best_solution[i])
```
