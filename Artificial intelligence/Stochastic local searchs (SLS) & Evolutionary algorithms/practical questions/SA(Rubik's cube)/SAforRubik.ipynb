```python
import numpy as np
import random
import math
import pandas as pd
import copy

def read_cube_faces():
    f=open('./Rubik\'s Cube Input.txt','r')
    up=[]
    down=[]
    left=[]
    right=[]
    front=[]
    back=[]
    for i in range(6):
        for j in range(3):
            if i==0:
                up.append(list(int(k) for k in f.readline().split()))
            elif i==1:
                left.append(list(int(k) for k in f.readline().split()))
            elif i==2:
                back.append(list(int(k) for k in f.readline().split()))
            elif i==3:
                right.append(list(int(k) for k in f.readline().split()))
            elif i==4:
                front.append(list(int(k) for k in f.readline().split()))
            else:
                down.append(list(int(k) for k in f.readline().split()))
        f.readline()

    return (up,left,back,right,front,down)

def time_temprature_schedule(time,T):
    return 1/(2*time+1)*T
    #return T/math.log2(2+time)

def SA_for_Rubik(faces):
# SA patameters
    T=10
    coefficient=0.999
    time=0
    initial_T=10


# initial solution
    current=faces
    current_value=0
    up,left,back,right,front,down=current
    for i in range(3):
        for j in range(3):
            if up[i][j]!=1:
                current_value+=1
            if left[i][j]!=2:
                current_value+=1
            if back[i][j]!=3:
                current_value+=1
            if right[i][j]!=4:
                current_value+=1
            if front[i][j]!=5:
                current_value+=1
            if down[i][j]!=6:
                current_value+=1

    
#Sa
    moves=[]
    while(1):
        #T*=coefficient
        T=time_temprature_schedule(time,initial_T)
        time+=1

        if T < 0.1 or current_value == 0:
            return current , current_value ,moves  # current_value == number of squares that are not in the right place it is must be 0 
        
        # construct a neighbor
        mat=[[0,0,1],[0,1,0],[1,0,0]]
        mat=np.array(mat)
        up,left,back,right,front,down=current
        up=np.array(up)
        left=np.array(left)
        back=np.array(back)
        right=np.array(right)
        front=np.array(front)
        down=np.array(down)

        rand_number=random.randint(1,12) # 1 = U , 2 = U' , 3 = L , 4 = L', 5 = B , 6 = B' , 7 = R , 8 = R' , 9 = F, 10 = F' , 11 = D , 12 = D'
        if rand_number == 1:
            up=np.dot(mat,up)
            up=up.T

            temp,front[0]=copy.deepcopy(front[0]),copy.deepcopy(right[0])
            temp,left[0]=copy.deepcopy(left[0]),copy.deepcopy(temp)
            temp,back[0]=copy.deepcopy(back[0]),copy.deepcopy(temp)
            right[0]=copy.deepcopy(temp)
            

        elif rand_number == 2:
            up=up.T
            up=np.dot(mat,up)

            temp,front[0]=copy.deepcopy(front[0]),copy.deepcopy(left[0])
            temp,right[0]=copy.deepcopy(right[0]),copy.deepcopy(temp)
            temp,back[0]=copy.deepcopy(back[0]),copy.deepcopy(temp)
            left[0]=copy.deepcopy(temp)


        elif rand_number == 3:
            left=np.dot(mat,left)
            left=left.T

            temp,down[:,0]=copy.deepcopy(down[:,0]),copy.deepcopy(back[:,0])
            temp,front[:,2]=copy.deepcopy(front[:,2]),copy.deepcopy(temp)
            temp,up[:,0]=copy.deepcopy(up[:,0]),copy.deepcopy(temp)
            back[:,0]=copy.deepcopy(temp)


        elif rand_number == 4:
            left=left.T
            left=np.dot(mat,left)

            temp,down[:,0]=copy.deepcopy(down[:,0]),copy.deepcopy(front[:,0])
            temp,back[:,2]=copy.deepcopy(back[:,2]),copy.deepcopy(temp)
            temp,up[:,0]=copy.deepcopy(up[:,0]),copy.deepcopy(temp)
            front[:,0]=copy.deepcopy(temp)


        elif rand_number == 5:
            back=np.dot(mat,back)
            back=back.T

            temp,down[0]=copy.deepcopy(down[0]),copy.deepcopy(right[:,0])
            temp,left[:,2]=copy.deepcopy(left[:,2]),copy.deepcopy(temp)
            temp,up[2]=copy.deepcopy(up[2]),copy.deepcopy(temp)
            right[:,0]=copy.deepcopy(temp)


        elif rand_number == 6:
            back=back.T
            back=np.dot(mat,back)

            temp,down[0]=copy.deepcopy(down[0]),copy.deepcopy(left[:,0])
            temp,right[:,2]=copy.deepcopy(right[:,2]),copy.deepcopy(temp)
            temp,up[2]=copy.deepcopy(up[2]),copy.deepcopy(temp)
            left[:,0]=copy.deepcopy(temp)


        elif rand_number == 7:
            right=np.dot(mat,right)
            right=right.T

            temp,down[:,2]=copy.deepcopy(down[:,2]),copy.deepcopy(back[:,2])
            temp,front[:,0]=copy.deepcopy(front[:,0]),copy.deepcopy(temp)
            temp,up[:,2]=copy.deepcopy(up[:,2]),copy.deepcopy(temp)
            back[:,2]=copy.deepcopy(temp)


        elif rand_number == 8:
            right=right.T
            right=np.dot(mat,right)

            temp,down[:,2]=copy.deepcopy(down[:,2]),copy.deepcopy(front[:,2])
            temp,back[:,0]=copy.deepcopy(back[:,0]),copy.deepcopy(temp)
            temp,up[:,2]=copy.deepcopy(up[:,2]),copy.deepcopy(temp)
            front[:,2]=copy.deepcopy(temp)


        elif rand_number == 9:
            front=np.dot(mat,front)
            front=front.T

            temp,down[2]=copy.deepcopy(down[2]),copy.deepcopy(right[:,2])
            temp,left[:,0]=copy.deepcopy(left[:,0]),copy.deepcopy(temp)
            temp,up[0]=copy.deepcopy(up[0]),copy.deepcopy(temp)
            right[:,2]=copy.deepcopy(temp)



        elif rand_number == 10:
            front=front.T
            front=np.dot(mat,front)

            temp,down[2]=copy.deepcopy(down[2]),copy.deepcopy(left[:,2])
            temp,right[:,0]=copy.deepcopy(right[:,0]),copy.deepcopy(temp)
            temp,up[0]=copy.deepcopy(up[0]),copy.deepcopy(temp)
            left[:,2]=copy.deepcopy(temp)


        elif rand_number == 11:
            down=np.dot(mat,down)
            down=down.T

            temp,front[2]=copy.deepcopy(front[2]),copy.deepcopy(right[2])
            temp,left[2]=copy.deepcopy(left[2]),copy.deepcopy(temp)
            temp,back[2]=copy.deepcopy(back[2]),copy.deepcopy(temp)
            right[2]=copy.deepcopy(temp)


        elif rand_number == 12:
            down=down.T
            down=np.dot(mat,down)

            temp,front[2]=copy.deepcopy(front[2]),copy.deepcopy(left[2])
            temp,right[2]=copy.deepcopy(right[2]),copy.deepcopy(temp)
            temp,back[2]=copy.deepcopy(back[2]),copy.deepcopy(temp)
            left[2]=copy.deepcopy(temp)



    # evaluation neighbor
        next_value=0
        for i in range(3):
            for j in range(3):
                if up[i][j]!=1:
                    next_value+=1
                if left[i][j]!=2:
                    next_value+=1
                if back[i][j]!=3:
                    next_value+=1
                if right[i][j]!=4:
                    next_value+=1
                if front[i][j]!=5:
                    next_value+=1
                if down[i][j]!=6:
                    next_value+=1

        #Delta E
        deltaE=next_value-current_value

        if deltaE<0:
            current=(up,left,back,right,front,down)
            current_value=next_value

            if rand_number == 1:
                moves.append('U')
            elif rand_number == 2:
                moves.append('U\'')
            elif rand_number == 3:
                moves.append('L')
            elif rand_number == 4:
                moves.append('L\'')
            elif rand_number == 5:
                moves.append('B')
            elif rand_number == 6:
                moves.append('B\'')
            elif rand_number == 7:
                moves.append('R')
            elif rand_number == 8:
                moves.append('R\'')
            elif rand_number == 9:
                moves.append('F')
            elif rand_number == 10:
                moves.append('F\'')
            elif rand_number == 11:
                moves.append('D')
            elif rand_number == 12:
                moves.append('D\'')

        else:
            if random.random()<= math.exp(-deltaE/T):
                current=(up,left,back,right,front,down)
                current_value=next_value

                if rand_number == 1:
                    moves.append('U')
                elif rand_number == 2:
                    moves.append('U\'')
                elif rand_number == 3:
                    moves.append('L')
                elif rand_number == 4:
                    moves.append('L\'')
                elif rand_number == 5:
                    moves.append('B')
                elif rand_number == 6:
                    moves.append('B\'')
                elif rand_number == 7:
                    moves.append('R')
                elif rand_number == 8:
                    moves.append('R\'')
                elif rand_number == 9:
                    moves.append('F')
                elif rand_number == 10:
                    moves.append('F\'')
                elif rand_number == 11:
                    moves.append('D')
                elif rand_number == 12:
                    moves.append('D\'')

        
best_solution,best_solution_value,moves=SA_for_Rubik(read_cube_faces())
up,left,back,right,front,down=best_solution

with open('./Rubik\'s Cube Output.txt','w') as f:
    for i in range(3):
        for j in range(3):
            f.write(f"{up[i][j]} ")
        f.write(f"\n")
    f.write(f"\n")

    for i in range(3):
        for j in range(3):
            f.write(f"{left[i][j]} ")
        f.write(f"\n")
    f.write(f"\n")

    for i in range(3):
        for j in range(3):
            f.write(f"{back[i][j]} ")
        f.write(f"\n")
    f.write(f"\n")

    for i in range(3):
        for j in range(3):
            f.write(f"{right[i][j]} ")
        f.write(f"\n")
    f.write(f"\n")

    for i in range(3):
        for j in range(3):
            f.write(f"{front[i][j]} ")
        f.write(f"\n")
    f.write(f"\n")

    for i in range(3):
        for j in range(3):
            f.write(f"{down[i][j]} ")
        f.write(f"\n")
    f.write(f"\n")

    for i in moves:
        f.write(f"{i}\n")


print("The value of objective function is (it must be 0)",best_solution_value)


       
```
