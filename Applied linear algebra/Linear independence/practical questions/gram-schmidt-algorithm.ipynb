```python
import numpy as np

def Gram_Schmidt_Algorithm(mat,rows,columns):

    Q=np.zeros_like(mat,dtype=np.float64) # q vectors
    C=np.zeros((columns,columns),dtype=np.float64)  # coefficients of dependent vectors
    D=[]     # indices of dependent vectors in matrix
    b=False
    for i in range(columns):
        V=mat[:,i] # vector i
        temp=[]
        for j in range(i):
            temp.append(np.dot(Q[:,j],V))
            
            V=V-temp[j]*Q[:,j]
           

        if np.all(V<0.0001):
            D.append(i)

            for i in range(columns-len(temp)):
                temp.append(0)
            C[:,i]=temp
            b=True
        else:
            Q[:,i]=V/(np.linalg.norm(V))
          
    if(b==False):
        print("Vectors are linearly independent.\nBasis vectors:")
        for i in range(columns):
            print("q"+str(i),": ",Q[:,i])

    else:
        print("Vectors are linearly dependent.\nindices of dependent vectors are(first index is 0):")
        
        print(D)
       
        for i in range(len(D)):
            print("Coefficients of vector ",D[i]," are: ",C[:,i]) # error
       




def get_vectors():
    rows=int(input("Enter number of rows:"))
    columns=int(input("Enter number of columns:"))
    mat=[]
    for i in range(rows):
        mat.append(list(int(j) for j in input().split()))

    np_mat=np.array(mat)
    

    Gram_Schmidt_Algorithm(np_mat,rows,columns)




get_vectors()
```
