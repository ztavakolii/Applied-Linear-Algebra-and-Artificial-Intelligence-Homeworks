```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

def show_image(data,title):
    image=data.reshape(40,30)
    plt.imshow(image , cmap='gray')
    plt.title(title)
    plt.show()

def read_data():
    data = pd.read_csv('./TrainData.txt',header=None,delimiter=',')
    data=data.values
    return data

def read_test_data():
    data = pd.read_csv('./TestData.txt',header=None,delimiter=',')
    data=data.values
    return data

```

```python
# a.I

data=read_data()

arrays_num=10
array_length=1200
random_centers=np.random.randint(2,size=(arrays_num,array_length))

#print(random_centers.shape)

for i in range(10):
    show_image(random_centers[i].reshape(40,30),'Initial Center '+str(i+1))

kmeans=KMeans(n_clusters=10,init=random_centers,n_init=1,max_iter=300,random_state=0)
kmeans.fit(data)

final_centers=kmeans.cluster_centers_
loss=kmeans.inertia_

print('The value of loss function is: ',loss)

for i in range(10):
    show_image(final_centers[i].reshape(40,30),'Final Center '+str(i+1))


#c
test_data=read_test_data()
labels=kmeans.predict(test_data)

for i in range(13):
    show_image(test_data[i].reshape(40,30),'Test data '+str(i+1))

for l in labels:
    show_image(final_centers[l].reshape(40,30),str(l))

```

```python
#a.II

data=read_data()

random_centers_from_data_indices=np.random.choice(data.shape[0],10,replace=False)
random_centers_from_data=data[random_centers_from_data_indices,:]

for i in range(10):
    show_image(random_centers_from_data[i].reshape(40,30),'Initial center '+str(i+1))

kmeans=KMeans(n_clusters=10,init=random_centers_from_data,n_init=1,max_iter=300,random_state=0)
kmeans.fit(data)

final_centers=kmeans.cluster_centers_
loss=kmeans.inertia_

print('The value of loss function is: ',loss)

for i in range(10):
    show_image(final_centers[i].reshape(40,30),'Final Center '+str(i+1))


#c
test_data=read_test_data()
labels=kmeans.predict(test_data)

for i in range(13):
    show_image(test_data[i].reshape(40,30),'Test data '+str(i+1))

for l in labels:
    show_image(final_centers[l].reshape(40,30),str(l))


```
